## 4. Data Cleaning

```python
import pandas as pd

def clean_trading_data_corrected(file_path):
    """
    Correctly reads and cleans the data by specifying all column names
    and correctly parsing the Time index.
    """
    try:
        # Define the headers for all 7 columns.
        column_names = ['Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Spread']
        
        # Read the CSV, skipping the original malformed header row (header=0)
        # and applying our defined names.
        # We also set 'Time' as the index and parse its dates immediately.
        df = pd.read_csv(
            file_path, 
            sep='\t', 
            header=0, 
            names=column_names, 
            index_col='Time', 
            parse_dates=True
        )

        # Convert numeric columns to ensure correct data types.
        numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'Spread']
        for col in numeric_cols:
            df[col] = pd.to_numeric(df[col], errors='coerce')

        # Drop rows that might have missing values.
        df.dropna(inplace=True)
        
        # Sort by time index, which is good practice.
        df.sort_index(inplace=True)

        print("Data cleaning completed successfully.")
        return df

    except Exception as e:
        print(f"An error occurred: {e}")
        return None

eurusd_D1_df = clean_trading_data_corrected('/Users/pawan/Desktop/Quantwater Tech Investments/Operations/Internship/IIT Kanpur MBA Capstone Project/qti-internship-2025/trading-automation/data/forex/EURUSD_D1.csv')
#eurusd_H1_df = clean_trading_data_corrected('/Users/pawan/Desktop/Quantwater Tech Investments/Operations/Internship/IIT Kanpur MBA Capstone Project/qti-internship-2025/trading-automation/data/forex/EURUSD_H1.csv')
#eurusd_M15_df = clean_trading_data_corrected('/Users/pawan/Desktop/Quantwater Tech Investments/Operations/Internship/IIT Kanpur MBA Capstone Project/qti-internship-2025/trading-automation/data/forex/EURUSD_M15.csv')

xauusd_D1_df = clean_trading_data_corrected('/Users/pawan/Desktop/Quantwater Tech Investments/Operations/Internship/IIT Kanpur MBA Capstone Project/qti-internship-2025/trading-automation/data/commodities/XAUUSD_D1.csv')
#xauusd_H1_df = clean_trading_data_corrected('/Users/pawan/Desktop/Quantwater Tech Investments/Operations/Internship/IIT Kanpur MBA Capstone Project/qti-internship-2025/trading-automation/data/commodities/XAUUSD_H1.csv')
#xauusd_M15_df = clean_trading_data_corrected('/Users/pawan/Desktop/Quantwater Tech Investments/Operations/Internship/IIT Kanpur MBA Capstone Project/qti-internship-2025/trading-automation/data/commodities/XAUUSD_M15.csv')

print(eurusd_D1_df.head())
# To run this corrected function:
# eurusd_df = clean_trading_data_corrected('path/to/EURUSD_D1.csv')
# print(eurusd_df.head())
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from datetime import datetime
import pandas as pd

def plot_candlestick_chart(df, title="Candlestick Chart", figsize=(15, 8)):
    """
    Create a candlestick chart from OHLC data.
    
    Parameters:
    df (DataFrame): DataFrame with columns ['Open', 'High', 'Low', 'Close']
    title (str): Chart title
    figsize (tuple): Figure size (width, height)
    """
    fig, ax = plt.subplots(figsize=figsize)
    
    # Ensure we have the required columns
    required_cols = ['Open', 'High', 'Low', 'Close']
    if not all(col in df.columns for col in required_cols):
        raise ValueError(f"DataFrame must contain columns: {required_cols}")
    
    # Get the data
    opens = df['Open'].values
    highs = df['High'].values
    lows = df['Low'].values
    closes = df['Close'].values
    
    # Create x-axis positions (assuming daily data)
    x_positions = range(len(df))
    
    # Plot candlesticks
    for i, (open_price, high, low, close) in enumerate(zip(opens, highs, lows, closes)):
        # Determine if it's a bullish (green) or bearish (red) candle
        color = 'green' if close >= open_price else 'red'
        
        # Draw the high-low line (wick)
        ax.plot([i, i], [low, high], color='black', linewidth=1)
        
        # Draw the open-close rectangle (body)
        body_height = abs(close - open_price)
        body_bottom = min(open_price, close)
        
        # Create rectangle for the body
        rect = patches.Rectangle((i - 0.4, body_bottom), 0.8, body_height, 
                               linewidth=1, edgecolor='black', facecolor=color, alpha=0.7)
        ax.add_patch(rect)
        
        # Draw open and close ticks
        ax.plot([i - 0.4, i], [open_price, open_price], color='black', linewidth=2)
        ax.plot([i, i + 0.4], [close, close], color='black', linewidth=2)
    
    # Customize the chart
    ax.set_title(title, fontsize=16, fontweight='bold')
    ax.set_xlabel('Time', fontsize=12)
    ax.set_ylabel('Price', fontsize=12)
    ax.grid(True, alpha=0.3)
    
    # Set x-axis labels (show every nth date to avoid crowding)
    step = max(1, len(df) // 10)  # Show about 10 labels
    ax.set_xticks(x_positions[::step])
    ax.set_xticklabels([df.index[i].strftime('%Y-%m-%d') for i in range(0, len(df), step)], 
                      rotation=45, ha='right')
    
    # Add legend
    from matplotlib.patches import Patch
    legend_elements = [Patch(facecolor='green', alpha=0.7, label='Bullish (Close ≥ Open)'),
                      Patch(facecolor='red', alpha=0.7, label='Bearish (Close < Open)')]
    ax.legend(handles=legend_elements, loc='upper left')
    
    plt.tight_layout()
    plt.show()

def plot_candlestick_subset(df, start_date=None, end_date=None, title="Candlestick Chart", figsize=(15, 8)):
    """
    Plot candlestick chart for a specific date range.
    
    Parameters:
    df (DataFrame): DataFrame with OHLC data
    start_date (str): Start date in 'YYYY-MM-DD' format
    end_date (str): End date in 'YYYY-MM-DD' format
    title (str): Chart title
    figsize (tuple): Figure size
    """
    # Filter data by date range if specified
    if start_date or end_date:
        if start_date:
            df = df[df.index >= start_date]
        if end_date:
            df = df[df.index <= end_date]
    
    plot_candlestick_chart(df, title, figsize)

# Example usage:
#plot_candlestick_chart(eurusd_D1_df, "EUR/USD Daily Candlestick Chart")
plot_candlestick_subset(eurusd_D1_df, "2025-01-01", "2025-03-31", "EUR/USD 2025 Q1 Candlestick Chart")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from datetime import datetime

def plot_correlation_heatmap(df_list, labels=None, title="Correlation Heatmap", figsize=(10, 8)):
    """
    Create a correlation heatmap for multiple datasets.
    
    Parameters:
    df_list (list): List of DataFrames with price data
    labels (list): Labels for each dataset
    title (str): Chart title
    figsize (tuple): Figure size
    """
    if labels is None:
        labels = [f"Dataset {i+1}" for i in range(len(df_list))]
    
    # Calculate returns for each dataset
    returns_data = {}
    for i, df in enumerate(df_list):
        if 'Close' in df.columns:
            returns_data[labels[i]] = df['Close'].pct_change().dropna()
        else:
            returns_data[labels[i]] = df.iloc[:, 0].pct_change().dropna()
    
    # Create correlation matrix
    returns_df = pd.DataFrame(returns_data)
    correlation_matrix = returns_df.corr()
    
    # Create heatmap
    plt.figure(figsize=figsize)
    sns.heatmap(correlation_matrix, 
                annot=True, 
                cmap='coolwarm', 
                center=0,
                square=True,
                fmt='.3f',
                cbar_kws={'label': 'Correlation Coefficient'})
    
    plt.title(title, fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    return correlation_matrix

def plot_returns_heatmap(df, title="Returns Heatmap", figsize=(12, 8)):
    """
    Create a heatmap of daily returns over time.
    
    Parameters:
    df (DataFrame): DataFrame with price data
    title (str): Chart title
    figsize (tuple): Figure size
    """
    # Calculate returns
    if 'Close' in df.columns:
        returns = df['Close'].pct_change().dropna()
    else:
        returns = df.iloc[:, 0].pct_change().dropna()
    
    # Create year-month matrix
    returns_df = pd.DataFrame({
        'Date': returns.index,
        'Returns': returns.values,
        'Year': returns.index.year,
        'Month': returns.index.month
    })
    
    # Pivot to create heatmap data
    heatmap_data = returns_df.pivot_table(
        values='Returns', 
        index='Year', 
        columns='Month', 
        aggfunc='mean'
    )
    
    # Create heatmap
    plt.figure(figsize=figsize)
    sns.heatmap(heatmap_data, 
                annot=True, 
                cmap='RdYlGn', 
                center=0,
                fmt='.4f',
                cbar_kws={'label': 'Average Daily Returns'})
    
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Year', fontsize=12)
    plt.tight_layout()
    plt.show()
    
    return heatmap_data

def plot_volatility_heatmap(df, title="Volatility Heatmap", figsize=(12, 8)):
    """
    Create a heatmap of volatility over time.
    
    Parameters:
    df (DataFrame): DataFrame with price data
    title (str): Chart title
    figsize (tuple): Figure size
    """
    # Calculate returns
    if 'Close' in df.columns:
        returns = df['Close'].pct_change().dropna()
    else:
        returns = df.iloc[:, 0].pct_change().dropna()
    
    # Create year-month matrix
    returns_df = pd.DataFrame({
        'Date': returns.index,
        'Returns': returns.values,
        'Year': returns.index.year,
        'Month': returns.index.month
    })
    
    # Pivot to create heatmap data (using std for volatility)
    heatmap_data = returns_df.pivot_table(
        values='Returns', 
        index='Year', 
        columns='Month', 
        aggfunc='std'
    )
    
    # Create heatmap
    plt.figure(figsize=figsize)
    sns.heatmap(heatmap_data, 
                annot=True, 
                cmap='Reds', 
                fmt='.4f',
                cbar_kws={'label': 'Volatility (Std Dev)'})
    
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Year', fontsize=12)
    plt.tight_layout()
    plt.show()
    
    return heatmap_data

def plot_price_heatmap(df, title="Price Heatmap", figsize=(12, 8)):
    """
    Create a heatmap of closing prices over time.
    
    Parameters:
    df (DataFrame): DataFrame with price data
    title (str): Chart title
    figsize (tuple): Figure size
    """
    # Get closing prices
    if 'Close' in df.columns:
        prices = df['Close']
    else:
        prices = df.iloc[:, 0]
    
    # Create year-month matrix
    prices_df = pd.DataFrame({
        'Date': prices.index,
        'Price': prices.values,
        'Year': prices.index.year,
        'Month': prices.index.month
    })
    
    # Pivot to create heatmap data
    heatmap_data = prices_df.pivot_table(
        values='Price', 
        index='Year', 
        columns='Month', 
        aggfunc='mean'
    )
    
    # Create heatmap
    plt.figure(figsize=figsize)
    sns.heatmap(heatmap_data, 
                annot=True, 
                cmap='viridis', 
                fmt='.2f',
                cbar_kws={'label': 'Average Price'})
    
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Year', fontsize=12)
    plt.tight_layout()
    plt.show()
    
    return heatmap_data

def plot_volume_heatmap(df, title="Volume Heatmap", figsize=(12, 8)):
    """
    Create a heatmap of trading volume over time.
    
    Parameters:
    df (DataFrame): DataFrame with volume data
    title (str): Chart title
    figsize (tuple): Figure size
    """
    if 'Volume' not in df.columns:
        print("No Volume column found in DataFrame")
        return None
    
    # Create year-month matrix
    volume_df = pd.DataFrame({
        'Date': df.index,
        'Volume': df['Volume'].values,
        'Year': df.index.year,
        'Month': df.index.month
    })
    
    # Pivot to create heatmap data
    heatmap_data = volume_df.pivot_table(
        values='Volume', 
        index='Year', 
        columns='Month', 
        aggfunc='mean'
    )
    
    # Create heatmap
    plt.figure(figsize=figsize)
    sns.heatmap(heatmap_data, 
                annot=True, 
                cmap='Blues', 
                fmt='.0f',
                cbar_kws={'label': 'Average Volume'})
    
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Year', fontsize=12)
    plt.tight_layout()
    plt.show()
    
    return heatmap_data

# Example usage:
plot_correlation_heatmap([eurusd_D1_df, xauusd_D1_df], ['EUR/USD', 'Gold'], "EUR/USD vs Gold Correlation")
plot_returns_heatmap(eurusd_D1_df, "EUR/USD Returns Heatmap")
plot_volatility_heatmap(eurusd_D1_df, "EUR/USD Volatility Heatmap")
plot_price_heatmap(eurusd_D1_df, "EUR/USD Price Heatmap")
plot_volume_heatmap(eurusd_D1_df, "EUR/USD Volume Heatmap")
# plot_correlation_heatmap([eurusd_fixed, xauusd_fixed], ['EUR/USD', 'Gold'], "EUR/USD vs Gold Correlation")
# plot_returns_heatmap(eurusd_fixed, "EUR/USD Returns Heatmap")
# plot_volatility_heatmap(eurusd_fixed, "EUR/USD Volatility Heatmap")
# plot_price_heatmap(eurusd_fixed, "EUR/USD Price Heatmap")
# plot_volume_heatmap(eurusd_fixed, "EUR/USD Volume Heatmap")
```

```python
# Trading patterns

# bulish and bearish trends

def plot_bullish_bearish_trends(df, ma_window=50):
    """
    Plot bullish (green) and bearish (red) trends for EURUSD
    based on price vs moving average.
    """
    df = df.copy()
    df['MA'] = df['Close'].rolling(window=ma_window).mean()
    df['Trend'] = df['Close'] > df['MA']  # True = Bullish, False = Bearish

    plt.figure(figsize=(12,6))
    plt.plot(df['Close'], color='black', label='Close Price')
    plt.plot(df['MA'], color='blue', linestyle='--', label=f'{ma_window}-Day MA')

    # Highlight bullish/bearish regions
    plt.fill_between(df.index, df['Close'], df['MA'], 
                     where=df['Trend'], color='green', alpha=0.3, label='Bullish')
    plt.fill_between(df.index, df['Close'], df['MA'], 
                     where=~df['Trend'], color='red', alpha=0.3, label='Bearish')

    plt.title("EURUSD Bullish & Bearish Trends")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.show()

# plot the bullish and bearish trends for eurusd_D1_df for specific timeperiod
plot_bullish_bearish_trends(eurusd_D1_df[eurusd_D1_df.index.year >= 2023])



# plot the bullish and bearish trends for xauusd_D1_df
#plot_bullish_bearish_trends(xauusd_D1_df)

```

```python
# volatility analysis

def volatility_analysis(df, window=20):
    """
    Volatility analysis of EURUSD:
    1. Rolling Std Dev (Historical Vol)
    2. ATR
    3. Bollinger Band Width
    """
    df = df.copy()
    
    # 1. Historical Volatility (std of returns)
    df['Returns'] = df['Close'].pct_change()
    df['Rolling_Std'] = df['Returns'].rolling(window).std()
    df['HistVol_Ann'] = df['Rolling_Std'] * np.sqrt(252)  # annualized
    
    # 2. ATR
    df['H-L'] = df['High'] - df['Low']
    df['H-C'] = abs(df['High'] - df['Close'].shift(1))
    df['L-C'] = abs(df['Low'] - df['Close'].shift(1))
    df['TR'] = df[['H-L','H-C','L-C']].max(axis=1)
    df['ATR'] = df['TR'].rolling(window).mean()
    
    # 3. Bollinger Band Width
    df['MA'] = df['Close'].rolling(window).mean()
    df['BB_Upper'] = df['MA'] + 2*df['Rolling_Std']
    df['BB_Lower'] = df['MA'] - 2*df['Rolling_Std']
    df['BB_Width'] = (df['BB_Upper'] - df['BB_Lower']) / df['MA']
    
    # --- Plotting ---
    fig, axs = plt.subplots(3, 1, figsize=(12,10), sharex=True)

    # Price & ATR (with ATR on RHS)
    ax1 = axs[0]
    ax2 = ax1.twinx()  # Create second y-axis on the right
    
    # Plot Close price on left axis
    line1 = ax1.plot(df.index, df['Close'], color='black', label='Close')
    ax1.set_ylabel('Close Price', color='black')
    ax1.tick_params(axis='y', labelcolor='black')
    
    # Plot ATR on right axis
    line2 = ax2.plot(df.index, df['ATR'], color='orange', label='ATR')
    ax2.set_ylabel('ATR', color='orange')
    ax2.tick_params(axis='y', labelcolor='orange')
    
    ax1.set_title("EURUSD Close & ATR")
    
    # Combine legends
    lines = line1 + line2
    labels = [l.get_label() for l in lines]
    ax1.legend(lines, labels, loc='upper left')

    # Historical Volatility
    axs[1].plot(df.index, df['HistVol_Ann'], color='blue')
    axs[1].set_title("Annualized Historical Volatility")
    axs[1].set_ylabel('Volatility')

    # Bollinger Band Width
    axs[2].plot(df.index, df['BB_Width'], color='purple')
    axs[2].set_title("Bollinger Band Width (Relative Volatility)")
    axs[2].set_ylabel('BB Width')
    axs[2].set_xlabel('Date')
    
    plt.tight_layout()
    plt.show()
    
    return df[['HistVol_Ann','ATR','BB_Width']]

# plot the volatility analysis for eurusd_D1_df for specific timeperiod
volatility_analysis(eurusd_D1_df[eurusd_D1_df.index.year >= 2023])

# plot the volatility analysis for xauusd_D1_df for specific timeperiod
#volatility_analysis(xauusd_D1_df[xauusd_D1_df.index.year >= 2023])
```

```python

def plot_trends_with_volatility(df, ma_window=50, atr_window=20):
    """
    Plot bullish/bearish trends with volatility regimes and ATR (RHS).
    """
    df = df.copy()
    
    # --- Moving Average Trend ---
    df['MA'] = df['Close'].rolling(ma_window).mean()
    df['Trend'] = df['Close'] > df['MA']  # True = Bullish, False = Bearish

    # --- ATR Calculation ---
    df['H-L'] = df['High'] - df['Low']
    df['H-C'] = abs(df['High'] - df['Close'].shift(1))
    df['L-C'] = abs(df['Low'] - df['Close'].shift(1))
    df['TR'] = df[['H-L','H-C','L-C']].max(axis=1)
    df['ATR'] = df['TR'].rolling(atr_window).mean()

    # Volatility regime threshold (median ATR)
    vol_threshold = df['ATR'].median()

    # --- Plot ---
    fig, ax1 = plt.subplots(figsize=(14,7))

    # Price & MA
    ax1.plot(df.index, df['Close'], color='black', linewidth=1.2, label='Close Price')
    ax1.plot(df.index, df['MA'], color='blue', linestyle='--', label=f'{ma_window}-Day MA')

    # Bullish/Bearish shading
    ax1.fill_between(df.index, df['Close'], df['MA'], 
                     where=df['Trend'], color='green', alpha=0.2, label='Bullish')
    ax1.fill_between(df.index, df['Close'], df['MA'], 
                     where=~df['Trend'], color='red', alpha=0.2, label='Bearish')

    # Volatility regime shading
    ax1.fill_between(df.index, ax1.get_ylim()[0], ax1.get_ylim()[1],
                     where=df['ATR'] > vol_threshold, color='grey', alpha=0.1,
                     label='High Vol Regime')
    ax1.fill_between(df.index, ax1.get_ylim()[0], ax1.get_ylim()[1],
                     where=df['ATR'] <= vol_threshold, color='blue', alpha=0.1,
                     label='Low Vol Regime')

    # --- ATR on RHS axis ---
    ax2 = ax1.twinx()
    ax2.plot(df.index, df['ATR'], color='goldenrod', linewidth=1.5, label='ATR (RHS)')
    ax2.set_ylabel("ATR", color='goldenrod')

    # Labels & Legends
    ax1.set_title("EURUSD Trends with Volatility Regimes & ATR", fontsize=14)
    ax1.set_xlabel("Date")
    ax1.set_ylabel("Price")
    ax1.grid(True)

    # Merge legends from both axes
    lines_1, labels_1 = ax1.get_legend_handles_labels()
    lines_2, labels_2 = ax2.get_legend_handles_labels()
    ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc="upper left")

    plt.show()

# plot the trends with volatility for eurusd_D1_df for specific timeperiod
plot_trends_with_volatility(eurusd_D1_df[eurusd_D1_df.index.year >= 2023])

# plot the trends with volatility for xauusd_D1_df for specific timeperiod
#plot_trends_with_volatility(xauusd_D1_df[xauusd_D1_df.index.year >= 2023])
```
